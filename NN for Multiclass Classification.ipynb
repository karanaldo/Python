```python
import numpy as np
import pandas as pd
```

```python
def relu(x):
    return np.maximum(0,x)

def softmax(x):
    exp_x = np.exp(x)
    return exp_x / np.sum(exp_x, axis= 1, keepdims = True)
```

```python
def forward(X, W1, W2):
    z1 = np.dot(X,W1)
    a1 = relu(z1)
    z2 = np.dot(a1, W2)
    y_hat = softmax(z2)
    return y_hat, a1, z1
```

```python
def compute_loss(y_hat, y):
    loss = -np.mean(y * np.log(y_hat))
    return loss
```

```python
def backprop(X, y, y_hat, a1, z1, W1, W2):
    m = X.shape[0]
    dL_dz2 = (y_hat - y) / m
    dL_dW2 = np.dot(a1.T, dL_dz2)
    dL_da1 = np.dot(dL_dz2, W2.T)
    dL_dz1 = dL_da1 * (z1 > 0)
    dL_dW1 = np.dot(X.T, dL_dz1)
    return dL_dW1, dL_dW2
```
