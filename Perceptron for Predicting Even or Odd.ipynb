```python
#Converting the number into Binary form

def binary_converter(n):
    binary = []
    while n > 0:
        binary.append(n % 2)
        n = n // 2
    binary.reverse()
    return binary

#Example
n = 8
binary_converter(n)
```

```python
#Accessing the last digit in the binary form and applying condition

def perceptron(x):
    binary = binary_converter(x)
    last_digit = binary[-1]
    if last_digit == 0:
        return "EVEN"
    else:
        return "ODD"
    
#Asking the user for a number and giving the output
x = int(input('Enter a Number :- '))
perceptron(x)
```

```python
#Perceptron with learning, bacpropogation of error and training the perceptron

import numpy as np

#Sigmoid activation function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

#Derivative of the Sigmoid activation function
def sigmoid_derivative(z):
    return sigmoid(z) * (1 - sigmoid(z))

#Conversion from integet to binary
def int_to_binary(x):
    binary = []
    for i in range(8):
        binary.append(x % 2)
        x //= 2
    binary.reverse()
    return np.array(binary)

#Perceptron function
def perceptron(x, w, b):
    z = np.dot(w, x) + b
    a = sigmoid(z)
    if a >= 0.5:
        return "odd"
    else:
        return "even"

#Function for training the perceptron using backpropogation
def train_perceptron(epochs, learning_rate):
    w = np.random.randn(8)
    b = np.random.randn()
    for i in range(epochs):
        for j in range(256):
            x = int_to_binary(j)
            y = 1 if j % 2 == 1 else 0
            z = np.dot(w, x) + b
            a = sigmoid(z)
            error = y - a
            dw = learning_rate * error * sigmoid_derivative(z) * x
            db = learning_rate * error * sigmoid_derivative(z)
            w += dw
            b += db
    return w, b

#Actual training (weight and bias updation) and printing the results
w, b = train_perceptron(epochs=100, learning_rate=0.1)
print(w,b)
print("------------------------------------------------")
for i in range(256):
    x = int_to_binary(i)
    prediction = perceptron(x, w, b)
    print("Input:", i, "Prediction:", prediction)
```
