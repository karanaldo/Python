```python
#Linear or Identity

import matplotlib.pyplot as plt
import numpy as np

x= np.arange(-10,10)
y=x
plt.plot(x,y,c='blue')
plt.title('Linear')
plt.show()
```

```python
#Binary or Step Function

x= np.arange(-10,10)
y=[]
for i in range (x.shape[0]):
    if (x[i]>0):
        y.append(1)
    else:
        y.append(0)
plt.plot(x,y,c='green')
plt.title('Binary')
plt.show()
```

```python
#Bipolar

x=np.arange(-10,10)
y=[]
for i in range(x.shape[0]):
    if (x[i]>0):
        y.append(1)
    else:
        y.append(-1)
plt.plot(x,y,c='red')
plt.title('Bipolar')
plt.show()
```

```python
#Sigmoid

x=np.arange(-10,10)
y=[]
for i in range(x.shape[0]):
    y.append(1/(1+np.exp(-x[i])))
plt.plot(x,y,c='blue')
plt.title('Sigmoid')
plt.show()
```

```python
#Tanh

x=np.arange(-10,10)
y=[]
for i in range(x.shape[0]):
    y.append((np.exp(x[i])-np.exp(-x[i]))/(np.exp(x[i])+np.exp(-x[i])))
plt.plot(x,y,c='green')
plt.title('Tanh')
plt.show()
```

```python
#Relu

x=np.arange(-10,10)
y=[]
for i in range(x.shape[0]): 
    y.append(np.maximum(x[i],0))
plt.plot(x,y,c='red')
plt.title('Relu')
plt.show()
```

```python
#Leaky Relu

x=np.arange(-10,10)
y=[]

for i in range (x.shape[0]):
    y.append(np.maximum(x[i],x[i]*0.1))
plt.plot(x,y,c='blue')
plt.title('Leaky Relu')
plt.show()
```

```python
#ELU

x=np.arange(-10,10)
y=[]
a=1
for i in range(x.shape[0]):
    if (x[i]>0):
        y.append(x[i])
    else:
        y.append((a*(np.exp(x[i])-1)))
        
plt.plot(x,y,c='green')
plt.title('ELU')
plt.show()
```

```python
#Softmax

#exp_x = np.exp(x)      return exp_x / np.sum(exp_x, axis=0)

x=np.arange(-10,10)
y=[]

for i in range(x.shape[0]):
    y.append((np.exp(x[i])/(np.sum(np.exp(x)))))
    
plt.plot(x,y,c='red')
plt.title('Softmax')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linear(x,y):
    for i in range(x.shape[0]):
        y.append(x[i])
```

```python
def binary(x,y):
    for i in range(x.shape[0]):
        if x[i]>=0:
            y.append(1)
        else:
            y.append(0)
```

```python
def bipolar(x,y):
    for i in range (x.shape[0]):
        if x[i]>=0:
            y.append(1)
        else:
            y.append(-1)
```

```python
def sigmoid(x,y):
    for i in range (x.shape[0]):
        y.append(1/(1+np.exp(-x[i])))
```

```python
def tanh(x,y):
    for i in range (x.shape[0]):
        y.append((np.exp(x[i]) - np.exp(-x[i]))/(np.exp(x[i]) + np.exp(-x[i])))
```

```python
def relu(x,y):
    for i in range (x.shape[0]):
        y.append(np.maximum(x[i],0))
```

```python
def leakyrelu(x,y):
    for i in range (x.shape[0]):
        y.append(np.maximum(x[i],x[i]*0.1))
```

```python
def paremeterisedrelu(x,y):
    for i in range (x.shape[0]):
        a =0.3101 #trainable value, not hardcoded
        y.append(np.maximum(x[i],x[i]*a))
```

```python
def elu(x,y):
    for i in range (x.shape[0]):
        if x[i]>0:
            y.append(x[i])
        else:
            y.append(a*(np.exp(x[i])-1))
```

```python
def softmax(x,y):
    for i in range (x.shape[0]):
        y.append((np.exp(x[i])/(np.sum(np.exp(x)))))
```

```python
x = np.arange(-10,10)
y = []

#calling the function
sigmoid(x,y)

plt.plot(x,y)
plt.title("Sigmoid")
plt.show()
```
