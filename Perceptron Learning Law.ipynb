# Perceptron Learning Law

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate random input data
np.random.seed(42)
x = np.random.randn(100, 2)

# Define target labels
y = np.zeros((100,))
y[x[:,0] + x[:,1] > 0] = 1
```

```python
# Define sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Define perceptron learning rule function
def perceptron(x, y, a, epochs):
    # Initialize weights and bias
    w = np.zeros((2,))
    b = 0

    # Iterate over epochs
    for epoch in range(epochs):
        # Iterate over input data
        for i in range(x.shape[0]):
            # Calculate predicted output
            z = np.dot(w, x[i]) + b
            y_pred = sigmoid(z)
            
            #Try using sigmoid_derivative also

            # Update weights and bias
            w += a * (y[i] - y_pred) * x[i]
            b += a * (y[i] - y_pred)

    return w, b

# Train perceptron
w, b = perceptron(x, y, 0.2, 100)
```

```python
# Plot decision boundary and data

plt.figure(figsize=(6,6))
plt.scatter(x[y==0,0], x[y==0,1], c='r', label='Class 0')
plt.scatter(x[y==1,0], x[y==1,1], c='b', label='Class 1')
x1 = np.linspace(-3, 3, 10)
x2 = -(w[0]*x1 + b) / w[1]
plt.plot(x1, x2, c='k', label='Decision boundary')
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Perceptron Learning Rule')
plt.legend()
plt.show()
```
